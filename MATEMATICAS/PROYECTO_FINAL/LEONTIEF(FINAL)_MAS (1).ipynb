##**ACERCA DEL MODELO INSUMO - PRODUCTO (LEONTIEF)**

----

En Agosto de 1936 Wassily Leontief publicó "Quantitative Input and Output relations in the Economy System of the United States"  en *The Review of Economic Statistics*, teoría por la cual sería galardonado con el Premio Nobel de Economía en 1973.

La versión "abierta" de su modelo tiene como propósito determinar la producción bruta por sector para satisfacer la demanda de un país entero. 

**VARIABLES DE ENTRADA**

* S1, S2, S3,..., Sn son las industrias o sectores 
* $q_{ij}$  es número de unidades producidas por la industria $S_j$  para la industria $S_i$   
*$d$ es la demanda final o demanda externa de productos (lo que consume el sector no productivo)
* X1,X2,X3,...Xn es la producción total de cada industria  S1, S2, S3,..., Sn

**FÓRMULAS**   
En lugar de $q_{ij}$  el modelo necesita $a_{ij}$  , el número de unidades producidas por la industria  $S_j$ necesarias para producir una unidad en la industria $S_i$ . Construimos entonces la matriz de consumo total.


\begin{equation}
A =
\begin{pmatrix}
a_{11}  & \cdots & a_{1n}\\
\vdots & & \vdots\\
a_{n1}  & \cdots & a_{nn}
\end{pmatrix}
\end{equation}

Para estimar el modelo abierto de Leontief se requiere que  A y d (vector de demanda externa)  sean diferentes de cero. 

\begin{equation}
A {\quad} y {\quad} d  \neq
\begin{pmatrix}
0  \\
\vdots \\
 0\\
\end{pmatrix}
\end{equation}


Leontief plantea entonces, un sistema de ecuaciones lineales equivalentes a la ecuación matricial: 

\begin{equation}
X = AX  + d 
\end{equation}

Se puede transformar esta ecuación de la siguente manera: 

\begin{equation}
I_nX - AX    = d 
\end{equation}
\begin{equation}   
(I_nX - A) X = d       
\end{equation}
\begin{equation}   
X = (I_nX - A)^{-1}d       
\end{equation}   


Si la inversa de la matriz $(I_nX - A)$ existe entonces $(I_nX - A)^{-1}d$ es llamada la inversa de Leontief.

```
# Importar bibliotecas a utilizar 

import pandas as pd 
import matplotlib as plt
import numpy as np 
import seaborn as sns 
```

```
# Cargar dataset (está almacenado en google drive )
data = pd.read_excel("/content/drive/My Drive/Classroom/METODOS MATEMATICOS PARA EL ANALISIS DE DATOS CUCEA/MIP 2013.xlsx") 
data.head()
```

```
#EDA 
data = data.drop(data.index[0])
data = data.set_index(['Unnamed: 0'])
data.head()
data = data.replace(np.nan,0)
data.info()
```

```
data.tail()
```

####**PROCEDIMIENTO**

```
# Matriz de Coeficientes (A)
# Calcular los requerimientos por cada dólar de producción 

matriz_coeficientes = data.loc[:,'11 - Agricultura, cría y explotación de animales, aprovechamiento forestal, pesca y caza':].div(data.iloc[-1]['11 - Agricultura, cría y explotación de animales, aprovechamiento forestal, pesca y caza':])
matriz_coeficientes.tail(1) # Comprobar que la división se efectó de forma correcta 
```

```
# Impacto COVID  : sujeto a consideración de un experto (ingresar un shok % en la demanda)

shock_porcentual = 1 #expresar en decimales (e.g. -.10)
data['11 - Agricultura, cría y explotación de animales, aprovechamiento forestal, pesca y caza'] = data['11 - Agricultura, cría y explotación de animales, aprovechamiento forestal, pesca y caza'] * shock_porcentual
```

```
# Quitar : valor aregado, insumo total, demanda final y producto total porque NO van en la matriz de coeficientes 
matriz_coeficientes = matriz_coeficientes.drop(['Demanda Final '],axis = 1)
matriz_coeficientes = matriz_coeficientes.drop(['Producto Total '] ,axis = 1) 
matriz_coeficientes = matriz_coeficientes.drop([matriz_coeficientes.index[-2] , matriz_coeficientes.index[-1]])  #Elimnar valor agregado y demanda final (son las últimas 2 filas)
```

```
#Estructura de la tabla 
rows = matriz_coeficientes.shape[0] 
cols = matriz_coeficientes.shape[1] 
print("Rows: " + str(rows)) 
print("Columns: " + str(cols)) 

matriz_coeficientes.tail()
```

```
# Matriz de coeficientes 
A =  np.matrix(matriz_coeficientes)
A
```

```
# Matriz de identidad (I)
I = np.identity(rows)
#matriz_identidad = pd.DataFrame(matriz_identidad)}
I = np.matrix(I)
I
```

```
# Matriz I-A 
matriz_diferencia = I - A 
matriz_diferencia
```

```
# Inversa de la matriz (I-A)
from numpy.linalg import inv
matriz_inversa  = np.linalg.inv(matriz_diferencia)
matriz_inversa
```

```
# Matriz de la demanda final (d)
demanda_final = data['Demanda Final ']
demanda_final = demanda_final.drop([demanda_final.index[-2] , demanda_final.index[-1]]) 
demanda_final
d = np.matrix(demanda_final)
d
```

```
# Matriz Leontief
matriz_leontief = d * matriz_inversa 
matriz_leontief 
```

```
# Presentar resultados por industria 
resultados = pd.DataFrame(matriz_leontief)  # Convertir en un dataframe
resultados.columns = matriz_coeficientes.columns  # Nombrar las industrias 
resultados = resultados.T  # Invertir la tabla para que se vea el nombre completo de la industria 
resultados.columns = ['Producción por industria (Millones de pesos a precios básicos)']  # Ponerle titulo a la columna
pd.options.display.float_format = '{:,.0f}'.format  # Cambiar el formato de número 
resultados 
```
